```python
import pandas as pd
```

```python
#load the dataset
df = pd.read_csv("the_burger_spot.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
#selecting columns
df[["Area","Item", "Date", "Orders"]]
```

```python
#Filtering on Conditions
df[df['Area'] == "Arcadia Bay"]
# df[df['Orders']>= 100]
```

```python
#filtering based on multiple conditions
df[(df['Category'] == "Burgers") & (df['Type of Order'] == "Takeaway")]
```

```python
#filter oy NULL Values
df[df['Item'].notnull()].head()
```

```python
#filtering based on first of string
df[df['Item'].str.startswith("B")].head(10)
```

```python
#Customer who bought burger
Burger_buyers= df[df['Category'] == "Burgers"]
Burger_buyers.sort_values(by="Orders", ascending=False).head(10)
```

```python
#Data Clenaing and Preparation
df = df.rename(columns={'Foods':'Categories'})
df.head(2)
```

```python
print(df)
```

```python
#Removing Outliers
df[df[('Orders')]< df[('Orders')].quantile(0.99)]
```

```python
#Group Variables
df.groupby('Categories')['Orders'].max()
```

```python
#Creating Pivot Table
df.pivot_table(index = 'Area',
               columns = 'Type of Order',
               values = 'Orders',
               aggfunc= 'mean')
```

```python
# Data Visualization
import matplotlib.pyplot as plt
```

```python
#Plotting histogram
df['Orders'].hist()
plt.title("Histogram of Orders")
plt.xlabel("Orders")
plt.ylabel("Frequency")
plt.show()
```

```python
#Creating a barchart
plt.figure(figsize = (10,3))
avg_order = df.groupby("Area")['Orders'].mean()
avg_order.plot(kind ="bar")
plt.title("Bar Chart of Total Orders in Each Area")
plt.xlabel("Area")
plt.ylabel("Average Orders")
```

```python
#Creating Pie Chart
plt.figure(figsize = (8,4))
order_count = df['Type of Order'].value_counts()
order_count.plot(kind="pie", autopct="%1.1f%%")
plt.title("Pie Chart of Types of Orders")
plt.show()
```
